```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('Au_nanoparticle_dataset.csv') 
```

```python
data_frame.head()
```

```python
# List of columns to drop
columns_to_drop = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
```

```python
# Create a new dataframe by dropping the specified columns
filtered_data_frame = data_frame.drop(columns=columns_to_drop)
```

```python
filtered_data_frame.head(20)
```

```python
# Calculate mean, standard deviation, and quartiles for each column
statistics = {}
for column in columns_to_drop:
    statistics[column] = {
        'mean': data_frame[column].mean(),
        'std_dev': data_frame[column].std(),
        '25%': data_frame[column].quantile(0.25),
        '50%': data_frame[column].quantile(0.50),
        '75%': data_frame[column].quantile(0.75)
    }
```

```python
stats_df = pd.DataFrame(statistics)
```

```python
# Transpose the dataframe for better readability
stats_df = stats_df.T
```

```python
# Display the statistics
print(stats_df)
```

```python
# Create a figure and a 1x4 layout of subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot a histogram for each feature
for i, column in enumerate(columns_to_drop):
    axes[i].hist(data_frame[column], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
filterd_data = data_frame[columns_to_drop]
```

```python
# Use Seaborn's pairplot to visualize the scatter plots and histograms
pairplot = sns.pairplot(filterd_data)

# Show the plot
pairplot.fig.suptitle("Pairplot of N_total, N_bulk, N_surface, and R_avg", y=1.02)
plt.show()
```

```python
# Create a PairGrid
g = sns.PairGrid(filterd_data)

# Map the plots
g.map_upper(sns.histplot)  # Bivariate histogram on the upper half
g.map_diag(sns.histplot, kde=True)  # Histogram with KDE on the diagonal
g.map_lower(sns.kdeplot)  # Bivariate KDE plot on the lower half

# Show the plot
g.fig.suptitle("Pairplot of N_total, N_bulk, N_surface, and R_avg", y=1.02)
plt.show()
```
